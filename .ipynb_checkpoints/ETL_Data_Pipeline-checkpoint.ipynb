```python
import pandas as pd
```

```python
df = pd.read_csv("tested.csv")
df
```

```python
df.info()
```

```python
# age value replace with  according pclass...
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.mean()))
```

```python
df
```

```python
#  Drop Irrelevant Columns:
df.drop(['Name','Cabin','Ticket'], axis=1, inplace = True)
```

```python
df
```

```python
# Convert Data Types:Convert the Sex column into numerical data for easier analysis.

df['Sex'] = df['Sex'].map({'male':1 , 'female':0})
```

```python
df
```

```python
# Feature Engineering:Create new features based on existing ones (e.g., Age group).


df['Age_group'] = pd.cut(df['Age'], bins=[0, 18, 60, 100], labels=["Child", "Adult", "Senior"])
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Correct way to fill missing values in the 'Fare' column
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())

```
